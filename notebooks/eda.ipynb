## Goal
Exploratory Data Analysis (EDA)
Understand the shape, quality, and characteristics of the dataset before modeling or processing.

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

# Load raw data
raw_path = Path("../ml/recommendation/user_item_interactions.csv")  # Adjust path if necessary
df = pd.read_csv(raw_path)
```

```python
# basic info
print('Dataset information:')
print(df.info())
print('Summary statistics:')
print(df.describe())
print('First few rows of the dataset:')
print(df.head())
```

```python
# Missing Values

df.isnull().sum()
sns.heatmap(df.isnull(), cbar=False)
```

```python
# Unique Values

print(df.nunique())
```

```python
# Distribution of Key Variables
# - Quantity / TotalAmount
# - Product count per user

sns.histplot(df["TotalQuantity"], bins=50)
```

```python
# Top Customers / Items

df.groupby("CustomerID")["TotalQuantity"].sum().sort_values(ascending=False).head(10)
df.groupby("StockCode")["TotalQuantity"].sum().sort_values(ascending=False).head(10)
```
